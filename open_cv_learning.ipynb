```python
#Arithmetic operation on Image
```

```python
#addition of images
#adding one image pixel with other image 
#addWeighed() function
img1 = cv2.imread("images/image3.jpg")
print("dimensions of image1",img1.shape)
img2 = cv2.imread("images/image2.jpg")
print("dimensions of image2",img2.shape)
dim = (2952,2952)
image1 = cv2.resize(img1,dim)
print("image1")
plt.imshow(image1)
plt.show()
image2 = cv2.resize(img2,dim)
print("image2")
plt.imshow(image2)
plt.show()
weighted_image = cv2.addWeighted(image1,0.4,image2,0.5,0)
print("\naddweighted_image")
plt.imshow(weighted_image)
plt.show()
```

```python
#subtraction of image
#subtraction of image pixels from a image
#subtract() function
s_img1 = cv2.imread("images/1bit1.png")
s_img2 = cv2.imread("images/2bit2.png")
print("image 1")
plt.imshow(s_img1)
plt.show()
print("image 2")
plt.imshow(s_img2)
plt.show()
subtracted_image = cv2.subtract(s_img1,s_img2)
print("output image")
plt.imshow(subtracted_image)
plt.show()
```

```python
#Bitwise AND operation
bitwise_and_image = cv2.bitwise_and(s_img1,s_img2)
print("output image")
plt.imshow(bitwise_and_image)
plt.show()
```

```python
#Bitwise OR operation
bitwise_or_image = cv2.bitwise_or(s_img1,s_img2)
print("output image")
plt.imshow(bitwise_or_image)
plt.show()
```

```python
#Bitwise XOR operation
bitwise_xor_image = cv2.bitwise_xor(s_img1,s_img2)
print("output image")
plt.imshow(bitwise_xor_image)
plt.show()
```

```python
##Bitwise NOT operation
bitwise_not_image = cv2.bitwise_not(bitwise_xor_image)
print("output image")
plt.imshow(bitwise_not_image)
plt.show()
```
